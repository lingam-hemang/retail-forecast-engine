```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv('../data/raw_data/Sample - Superstore (Complete).csv')
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed')
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
print(df.columns)
drop_cols = ['Unnamed: 0','Row ID', 'Order ID', 'Order Date', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State', 'Postal Code', 'Region', 'Product ID', 'Quantity', 'Discount', 'Profit']
df.drop(drop_cols,axis=1,inplace=True)
features = ['Category', 'Sub-Category', 'Product Name', 'Year', 'Month']
df_summed_agg = df.groupby(features)['Sales'].sum().reset_index()
# df.groupby().agg(Total_Sales=('Sales', 'sum'))
df_summed_agg

```

```python
from scipy.stats import skew

# Assuming your DataFrame is called df
sales_skew = skew(df_summed_agg['Sales'].dropna())
print(f"Skewness: {sales_skew:.2f}")
```
